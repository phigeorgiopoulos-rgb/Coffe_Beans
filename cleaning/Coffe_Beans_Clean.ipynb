```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time
from datetime import datetime

orders_raw=pd.read_excel('Raw Data.xlsx',sheet_name='orders')
orders=orders_raw[['Order ID','Order Date','Customer ID','Product ID','Quantity']]
customers=pd.read_excel('Raw Data.xlsx',sheet_name='customers')
products=pd.read_excel('Raw Data.xlsx',sheet_name='products')
```

```python
orders1=pd.merge(orders, customers, on='Customer ID', how='left')
orders2=pd.merge(orders1, products, on='Product ID', how='left')
orders2['Weekday']=orders2['Order Date'].dt.day_name()
orders2['Month']=orders2['Order Date'].dt.month
orders2['Year']=orders2['Order Date'].dt.year
orders2.replace({'Roast Type': {'L': 'Light', 'M': 'Medium', 'D': 'Dark'},
                 'Coffee Type': {'Ara': 'Arabica', 'Rob': 'Robusta', 'Exc': 'Excelsa', 'Lib': 'Liberica'}
}, inplace=True)
orders2.drop(columns='Price per 100g', axis=1, inplace=True)
orders2['Revenue']=orders2['Quantity']*orders2['Unit Price']
```

```python
orders_final=orders2.copy()
orders_final.to_csv('orders_clean.csv', index=False)
```
